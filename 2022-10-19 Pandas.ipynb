### Merge these two DataFrames into a single DataFrame called census.

Please use the following dataset for merging purposes:

- https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/state-populations.csv

- https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/census-divisions.csv

```python
!ls
```

```python
state_pop = pd.read_csv('state-populations.csv')

print(f'\nThe dimensions of "state-populations.csv" is {state_pop.shape}.')
state_pop[0:2]
```

```python
census_div = pd.read_csv('census-divisions.csv')

print(f'\nThe dimensions of "census-divisions.csv" is {census_div.shape}.')
census_div[0:2]
```

```python
census = pd.merge(state_pop, census_div, on = "state", how = "outer")

print(f'\nThe dimensions of "census" dataframe is {census.shape}.')
census[0:2]
```

### Re-shape census so that one column contains all population measures, and another the year attributes.

```python
shaped_census = census.melt(id_vars=['state','region','division'],var_name='year',value_name='population')
shaped_census.head()
```

```python
shaped_census.tail()
```

### Group the data by year and summarize it.

```python
shaped_census.groupby('year').describe()
```

### Group the data by region, division and year and summarize it.

```python
shaped_census.groupby(['region','division','year']).describe()
```
