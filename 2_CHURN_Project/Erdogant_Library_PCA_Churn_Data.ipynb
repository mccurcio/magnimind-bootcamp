# PCA of Churn Data Using Erdogant-PCA-Library

- For more information see: [Erdogant_Library_PCA](https://erdogant.github.io/pca/pages/html/index.html)

```python
import pandas as pd
import numpy as np

from pca import pca
```

```python
filename = "mcc_clean_churn.csv"

df = pd.read_csv(filename)

df.drop(['Unnamed: 0', 'Churned'], axis=1, inplace=True)

#df.columns
```

```python
col_labels = ['Act_Len', 'Int_Plan', 'VM_Plan', 'Num_VM', 
              'T_D_Min', 'T_D_Calls', 'T_D_Cost', 
              'T_E_Min', 'T_E_Calls', 'T_E_Cost', 
              'T_Nt_Min', 'T_Nt_Calls', 'T_Nt_Cost', 
              'T_Int_Min', 'T_Int_Calls', 'T_Int_Cost',
              'Num_Srv_Calls']

# Initialize pca
model = pca(normalize=True)

# Fit transform & include the column labels
results = model.fit_transform(df, col_labels=col_labels)

# Make plot with parameters: set cmap to None and label and legend to False. 
# Only Vectors will be plotted.
model.biplot(cmap=None, label=None, legend=False)
```

**NOTE 6:** It is interesting to note that the cost and min for the 4 time slots are parallel, indicating that they are collinear and one can be removed.
